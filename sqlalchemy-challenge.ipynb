```python
%matplotlib inline
from matplotlib import style
style.use('fivethirtyeight')
import matplotlib.pyplot as plt
```

```python
import numpy as np
import pandas as pd
import datetime as dt
```

```python
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func, inspect
```

```python
engine = create_engine("sqlite:///Resources/hawaii.sqlite")
```

```python
inspector = inspect(engine)
inspector.get_table_names()
```

```python
columns = inspector.get_columns('station')
for c in columns:
    print(c['name'], c["type"])
```

```python
Base = automap_base()
Base.prepare(autoload_with=engine)
```

```python
Base.classes.keys()
```

```python
Measurement = Base.classes.measurement
Station = Base.classes.station
```

```python
session = Session(engine)
```

## Exploratory Precipitation Analysis

```python
session.query(Measurement.date).order_by(Measurement.date.desc()).first()
```

```python
query_date = dt.date(2017, 8, 23) - dt.timedelta(days=365)
print(query_date)
```

```python
query_12months = session.query(Measurement.date, Measurement.prcp).\
    filter(Measurement.date >= query_date).all()
query_12months
```

```python
precipitation_df = pd.DataFrame(query_12months, columns=['date', 'prcp'])
print(precipitation_df.head())
```

```python
precipitation_df = precipitation_df.sort_values('date')
```

```python
precipitation_df.plot(x='date', y='prcp', rot=90)

plt.xlabel("Date")
plt.ylabel("Inches")

plt.show
```

```python
precipitation_df.describe()
```

```python
session.query(Station.id).distinct().count()
```

```python
session.query(Measurement.station, func.count(Measurement.station)).\
    group_by(Measurement.station).\
    order_by(func.count(Measurement.station).desc()).all()
```

```python
most_active_station = 'USC00519281'
active_station = session.query(
    func.min(Measurement.tobs),
    func.max(Measurement.tobs),
    func.avg(Measurement.tobs)
).filter(Measurement.station == most_active_station).all()
active_station
```

```python
active_station_query_12months = session.query(Measurement.tobs).\
    filter(Measurement.station == most_active_station, Measurement.date >= query_date).all()
active_station_query_12months
```

```python
temperature_data_df = pd.DataFrame(active_station_query_12months) #columns=['tobs', 'prcp'])
print(temperature_data_df.head())
```

```python
temperature_data_df.plot.hist(bins=12, figsize=(10, 6))

plt.xlabel('Temperature (F)')
plt.ylabel('Frequency')
plt.grid()
plt.show()
```
